## 2. 绘制散点图 

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
plt.figure(figsize=(10,7))
n = 2000
x = np.random.rand(1,n)
y = np.random.rand(1,n)
t = np.arctan2(x,y)
plt.scatter(x,y,c=t,s=5,alpha=0.5,marker='o')
plt.show()
```

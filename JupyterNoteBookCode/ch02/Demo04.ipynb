```python
'''
x = A  sin(a*t+n/2)
y = B  sin(b*t)
'''
from matplotlib import pyplot as plt
a = 9
b = 8
# 从-pi到pi上均匀分布201个点参数
t = np.linspace(-np.pi,np.pi,201)

x = np.sin(a * t + np.pi/2)
y = np.sin(b * t)
plt.plot(x,y)
plt.show()
```

**扩散图**

```python
points = np.arange(-5,5,0.01)
x_arr1,y_arr1 = np.meshgrid(points,points)
x_arr1
```

```python
z_arr = np.sqrt(x_arr1 ** 2 + y_arr1 **2)
z_arr
```

```python
from matplotlib.font_manager import FontProperties
font_set = FontProperties(fname=r'C:\Windows\Fonts\simkai.ttf',size=16)
plt.imshow(z_arr,cmap=plt.cm.Reds)
plt.colorbar()
plt.title("X与Y轴2次方矢量化扩散图",FontProperties=font_set)
plt.show()
```

```python
arrone = np.array([1.1,1.2,1.3,1.4,1.5])
arrtwo = np.array([2.1,2.2,2.3,2.4,2.5])
arrbool = ([True,False,True,True,False])
```

```python
result = [(x if c else y)
             for x,y,c in zip(arrone,arrtwo,arrbool)]
result
```

```python
res = np.where(arrbool,arrone,arrtwo)
res
```

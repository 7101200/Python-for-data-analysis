```python
import numpy as np
from pandas import Series,DataFrame
import pandas as pd
```

```python
df1 = DataFrame({'key':list('bbacaab'),'data1': range(1,8)})
df2 = DataFrame({'key':list('abd'),'data2': range(1,4)})
df1
```

```python
df2
```

```python
pd.merge(df1,df2)
```

```python
pd.merge(df1,df2,on='key')
```

```python
df3 = DataFrame({'lkey':list('bbacaab'),'data1': range(1,8)})
df4 = DataFrame({'rkey':list('abd'),'data2': range(1,4)})
df3
```

```python
df4
```

```python
pd.merge(df3,df4,left_on='lkey',right_on='rkey')
```

```python
df5 = DataFrame({'key':list('bbacab'),'data1': range(1,7)})
df6 = DataFrame({'key':list('ababd'),'data2': range(1,6)})
df5
```

```python
df6
```

```python
pd.merge(df5,df6,how='inner')
```

```python
left = DataFrame({'key1':['foo','foo','bar'],
                 'key2':['one','two','three'],
                 'lval':[1,2,3]})
right = DataFrame({'key1':['foo','foo','bar','bar'],
                 'key2':['one','one','one','two'],
                 'rval':[4,5,6,7]})
```

```python
pd.merge(left,right,on=['key1','key2'],how='outer')
```

```python
pd.merge(left,right,on='key1')
```

```python
pd.merge(left,right,on='key1',suffixes=('_left','_right'))
```

```python
pd.concat({'level1':df1,'leve12':df2},axis=1)
```

```python
pd.concat([df1,df2],axis=1,keys=['level1','level2'],names=['upper','lower'])
```

```python
restaurant = pd.read_csv('RestaurantData.csv',encoding='gbk')
num_counts = pd.crosstab(restaurant['周期'],restaurant['人数'])
num_counts = num_counts.loc[:,2:5]
# 格式化成"和为1"
num_pcts = num_counts.div(num_counts.sum(1).astype(float),axis=0)
num_pcts.plot(kind='bar',figsize=(10,7),fontsize=16,rot=360,stacked=True)
plt.show()
```

```python
def signals(x,y):
    return (1-(x ** 2 + y ** 2)) * np.exp(-y ** 3 /3)

x = np.arange(-1.5,1.5,0.1)
y = x

X,Y = np.meshgrid(x,y)
Z = signals(X,Y)

N = np.arange(-1,1.5,0.3)

cs = plt.contour(Z,N, linewidths=2, cmap=mpl.cm.jet)
plt.clabel(cs,inline=True,fmt='%1.1f',fontsize=12)
plt.colorbar(cs)

plt.title('等高线图')
plt.show()
```

```python
sap = 25
x = np.random.normal(5,1,sap)
y = np.random.normal(3,.5,sap)

fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111,projection='3d')
hist,xed,yed = np.histogram2d(x,y,bins=10)
elements = (len(xed) -1) * (len(yed) -1)
xpos,ypos = np.meshgrid(xed[:-1]+.25,yed[:-1]+.25)

xpos = xpos.flatten()
ypos = ypos.flatten()
zpos = np.zeros(elements)

dx = .1 * np.ones_like(zpos)
dy = dx.copy()
dz = hist.flatten()

ax.bar3d(xpos,ypos,zpos,dx,dy,dz,color='b',alpha=0.4)
ax.set_xlabel('X轴')
ax.set_ylabel('Y轴')
ax.set_zlabel('Z轴')

plt.show()
```

```python
from numpy.random import randn
```

```python
fig = plt.figure(figsize=(10,7))
ax1 = fig.add_subplot(221)
plt.plot(randn(50).cumsum(),'y--')
ax2 = fig.add_subplot(222)
ax2.hist(randn(100),bins=20,color='#0000ff',alpha=0.4)
ax3 = fig.add_subplot(223)
ax3.scatter(np.arange(60),np.arange(60) + 3 * randn(60))
ax4 = fig.add_subplot(224)
ax4.plot(randn(50).cumsum(),'r')
plt.show()
```

```python
row = 2
col = 3
fig,axs = plt.subplots(row,col,sharex=True,sharey=True,figsize=(10,7))
for i in range(row):
    for j in range(col):
        axs[i,j].hist(randn(800),bins=80,color='#ff0000',alpha=0.5)
plt.subplots_adjust(wspace=0,hspace=0.02)
plt.show()
```

```python
fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111)
t = np.arange(0.0,5.0,0.01)
s = np.cos(2*np.pi*t)
ax.plot(t,s,color='b',alpha=0.35)
ax.annotate('标注文本',xy=(2,1),xytext=(2.5,1.5),arrowprops=dict(color='r'),fontsize=18)
ax.set_ylim(-2,2)
plt.show()
```

```python
from datetime import datetime
import pandas as pd
```

```python
fig = plt.figure(figsize=(14,9))
ax = fig.add_subplot(111)

data = pd.read_csv('StockData.csv',index_col=0,parse_dates=True,encoding='gbk')
strok = data['收盘价']

strok.plot(ax=ax,color='b',alpha=0.5)

key_data = [
    (datetime(2007,10,11),'股市顶峰值'),
    (datetime(2008,3,12),'贝尔斯登银行衰退'),
    (datetime(2008,9,15),'雷曼兄弟破产')
]

for date, label in key_data:
    ax.annotate(label,xy =(date,strok.asof(date) + 70),
                xytext=(date,strok.asof(date) + 240),
                arrowprops=dict(color='r'),fontsize=16)

ax.set_title('2008-2009年世界金融危机的重要日期',fontsize=20)
ax.set_xlim(['5/1/2007','1/1/2010'])
ax.set_ylim([600,2000])
plt.show()
```
